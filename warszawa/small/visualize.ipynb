```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_one(data):
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

    ax1.plot(data.time, data.wait)
    ax1.set_ylabel('Cummulative wait time')
    
    ax2.plot(data.time, data.queue_len)
    ax2.set_ylabel('Cummulative queue length')
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_two(n1, d1, n2, d2):
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

    ax1.plot(d1.time, d1.wait, label=n1)
    ax1.plot(d2.time, d2.wait, label=n2)
    ax1.legend()
    ax1.set_ylabel('Cummulative wait time')
    
    ax2.plot(d1.time, d1.queue_len, label=n1)
    ax2.plot(d2.time, d2.queue_len, label=n2)
    ax2.legend()
    ax2.set_ylabel('Cummulative queue length')
    
    plt.tight_layout()
    plt.show()
```

```python
normal = pd.read_csv("baseline.csv")
plot_one(normal)
```

```python
ac = pd.read_csv("ac-test.csv")
plot_one(ac)
```

```python
plot_two("normla", normal, "ac", ac)
```
